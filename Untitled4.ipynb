```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import numpy as np
import matplotlib.pyplot as plt

# Define input widgets
num1 = widgets.FloatText(description="Num 1:")
num2 = widgets.FloatText(description="Num 2:")
output = widgets.Output()
plot_output = widgets.Output()

# Define function buttons
add_button = widgets.Button(description="Add")
sub_button = widgets.Button(description="Subtract")
mul_button = widgets.Button(description="Multiply")
div_button = widgets.Button(description="Divide")
sin_button = widgets.Button(description="Sine")
cos_button = widgets.Button(description="Cosine")
tan_button = widgets.Button(description="Tangent")
plot_button = widgets.Button(description="Plot Graph")

# Define functions for calculator operations
def add_numbers(b):
    with output:
        clear_output()
        result = num1.value + num2.value
        print(f"Result: {result}")

def subtract_numbers(b):
    with output:
        clear_output()
        result = num1.value - num2.value
        print(f"Result: {result}")

def multiply_numbers(b):
    with output:
        clear_output()
        result = num1.value * num2.value
        print(f"Result: {result}")

def divide_numbers(b):
    with output:
        clear_output()
        if num2.value != 0:
            result = num1.value / num2.value
            print(f"Result: {result}")
        else:
            print("Error: Division by zero")

def sine_number(b):
    with output:
        clear_output()
        result = np.sin(np.radians(num1.value))
        print(f"Sine of {num1.value}°: {result}")

def cosine_number(b):
    with output:
        clear_output()
        result = np.cos(np.radians(num1.value))
        print(f"Cosine of {num1.value}°: {result}")

def tangent_number(b):
    with output:
        clear_output()
        result = np.tan(np.radians(num1.value))
        print(f"Tangent of {num1.value}°: {result}")

# Define function for plotting
def plot_graph(b):
    with plot_output:
        clear_output()
        if num1.value < num2.value:
            x = np.linspace(num1.value, num2.value, 100)
            y = x ** 2  # Sample function (y = x^2)
            plt.plot(x, y, label="y = x^2")
            plt.xlabel("x")
            plt.ylabel("y")
            plt.title("Plot of y = x^2")
            plt.legend()
            plt.grid(True)
            plt.show()
        else:
            print("Please ensure Num 1 is less than Num 2 for plotting.")

# Assign functions to button clicks
add_button.on_click(add_numbers)
sub_button.on_click(subtract_numbers)
mul_button.on_click(multiply_numbers)
div_button.on_click(divide_numbers)
sin_button.on_click(sine_number)
cos_button.on_click(cosine_number)
tan_button.on_click(tangent_number)
plot_button.on_click(plot_graph)

# Display widgets
display(num1, num2)
display(add_button, sub_button, mul_button, div_button)
display(sin_button, cos_button, tan_button)
display(plot_button)
display(output, plot_output)
```
